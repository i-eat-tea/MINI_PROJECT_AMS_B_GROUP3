```python
import pandas as pd
import numpy as np

# ── 1. Load ───────────────────────────────────────────────────────────────────
print(f"Original shape: {df.shape}")
df = df.copy()  # defragment on load

# ── 2. Standardise column names ───────────────────────────────────────────────
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r"[\s\-]+", "_", regex=True)
    .str.replace(r"[^\w]", "", regex=True)
)

# ── 3. Reconstruct repeaters_total / repeaters_girl from grade-level cols ─────
# Private sector left these as NaN. Grade-level cols are fully populated for
# all rows. Summing G1–12 matches the public values exactly (verified).
grade_rep_total = [f"grade_{i}_repeaters_total" for i in range(1, 13)]
grade_rep_girl  = [f"grade_{i}_repeaters_girl"  for i in range(1, 13)]
df["repeaters_total"] = df[grade_rep_total].sum(axis=1)
df["repeaters_girl"]  = df[grade_rep_girl].sum(axis=1)
print("Reconstructed: repeaters_total and repeaters_girl from grade 1–12 sums")

# ── 4. Fix the private-sector NT_Primary mislabel ────────────────────────────
# Your rename_map had: "Female_Primary" -> "NT_Primary"  ← WRONG
#   NT_Primary     = Non-Teaching Staff with primary-level qualification
#   Female_Primary = Female teaching staff with primary-level qualification
# Fix: split into correct columns before any drops

if "nt_primary" in df.columns:
    mask_priv = df["type"] == "private"
    # female_primary: private rows hold the mislabelled value; public = NaN
    df["female_primary"] = np.where(mask_priv, df["nt_primary"], np.nan)
    # nt_primary: correct to NaN for private (that data was never collected)
    df["nt_primary"]     = np.where(mask_priv, np.nan, df["nt_primary"])
    print("Fixed: private nt_primary → female_primary (wrong rename_map undone)")

# ── 5. Drop high-NaN and unnecessary columns ──────────────────────────────────
# 35 columns removed:
#   total_enrollment_*/total_repeaters_* — aggregate duplicates (now reconstructed)
#   disadv_schools / classes_pagoda      — requested removal
#   total_postgrad / total_phd           — all-null
#   t_postgrad / t_phd                   — only public has these (50%+ NaN)
#   nt_lsec–nt_phd                       — not collected for private (50%+ NaN)
#   noped_lsec–noped_phd                 — not collected for private (50%+ NaN)
#   female_lsec–female_phd               — only private trailing cols (50%+ NaN)
#   nopedtrain_*                         — only private trailing cols (50%+ NaN)
#   foreigner_staff_*                    — only public (50%+ NaN)
#   buildings / rooms / classrooms       — only private trailing cols (50%+ NaN)

drop_cols = [
    "total_enrollment_total", "total_enrollment_girl",
    "total_repeaters_total",  "total_repeaters_girl",
    "disadv_schools",         "classes_pagoda",
    "total_postgrad",         "total_phd",
    "t_postgrad",             "t_phd",
    "nt_lsec",   "nt_usec",  "nt_graduate", "nt_postgrad", "nt_phd",
    "noped_lsec","noped_usec","noped_graduate","noped_postgrad","noped_phd",
    "female_lsec","female_usec","female_graduate","female_postgrad","female_phd",
    "nopedtrain_lsec","nopedtrain_usec","nopedtrain_graduate",
    "nopedtrain_postgrad","nopedtrain_phd",
    "foreigner_staff_total","foreigner_staff_female",
    "buildings","rooms","classrooms",
]
drop_cols = [c for c in drop_cols if c in df.columns]
df.drop(columns=drop_cols, inplace=True)
print(f"Dropped {len(drop_cols)} columns")

# ── 6. Fix data types ─────────────────────────────────────────────────────────
df["year"]     = pd.to_numeric(df["year"],     errors="coerce").astype("Int64")
df["type"]     = df["type"].astype(str).str.strip()
df["province"] = df["province"].astype(str).str.strip()
id_cols = {"type", "province", "year"}
for col in df.columns:
    if col not in id_cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Defragment to clean up any internal fragmentation warnings
df = df.copy()

# ── 7. Remove duplicate rows ──────────────────────────────────────────────────
before = len(df)
df.drop_duplicates(inplace=True)
if before - len(df):
    print(f"Removed {before - len(df)} duplicate rows")

# ── 8. Sort ───────────────────────────────────────────────────────────────────
df.sort_values(["type", "province", "year"], inplace=True, ignore_index=True)

# ── 9. Final NaN audit ────────────────────────────────────────────────────────
print(f"\nFinal shape: {df.shape}")
nan_pct = (df.isna().sum() / len(df) * 100).sort_values(ascending=False)
remaining = nan_pct[nan_pct > 0]
if remaining.empty:
    print("No remaining NaN values.")
else:
    print(f"\nRemaining NaN (these are structural — data never collected for that type):")
    print(remaining.to_string())

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visual style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

# ------------------------------
# 1. Load the data
# ------------------------------

# If the data is already in a DataFrame, you can skip the line above
# and directly use the variable 'df' containing the data.

# ------------------------------
# 2. Basic information
# ------------------------------
print("Shape of the dataset:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nColumn names:")
print(df.columns.tolist())
print("\nData types:")
print(df.dtypes)

# ------------------------------
# 3. Missing values
# ------------------------------
missing = df.isnull().sum()
missing_pct = 100 * missing / len(df)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print("\nMissing values per column:")
print(missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False))

# ------------------------------
# 4. Summary statistics for numeric columns
# ------------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("\nSummary statistics for numeric columns:")
print(df[numeric_cols].describe().T)

# ------------------------------
# 5. Distribution of categorical variables
# ------------------------------
categorical_cols = ['type', 'province', 'year']
for col in categorical_cols:
    if col in df.columns:
        print(f"\nValue counts for '{col}':")
        print(df[col].value_counts())

# ------------------------------
# 6. Univariate analysis: histograms for key numeric features
# ------------------------------
key_numeric = ['enrollment_total', 'teachstaff_total', 'totalstaff_total',
               'grade_1_enrollment_total', 'grade_12_enrollment_total',
               't_primary', 't_lsec', 't_usec', 't_graduate']

# Filter only columns that exist in the dataset
key_numeric = [col for col in key_numeric if col in df.columns]

if key_numeric:
    df[key_numeric].hist(bins=30, figsize=(15, 12))
    plt.suptitle('Distributions of Key Numeric Variables', fontsize=16)
    plt.tight_layout()
    plt.show()

# ------------------------------
# 7. Time series analysis: average enrollments over years
# ------------------------------
if 'year' in df.columns and 'enrollment_total' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x='year', y='enrollment_total', hue='type', ci=None)
    plt.title('Total Enrollment over Years by School Type')
    plt.ylabel('Total Enrollment')
    plt.show()

# ------------------------------
# 8. Comparison between public and private schools
# ------------------------------
if 'type' in df.columns:
    type_group = df.groupby('type')[['enrollment_total', 'teachstaff_total', 'totalstaff_total']].mean()
    print("\nAverage values by school type:")
    print(type_group)

    # Boxplot of enrollment by type
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='type', y='enrollment_total')
    plt.title('Enrollment Distribution by School Type')
    plt.show()

# ------------------------------
# 9. Province-level analysis
# ------------------------------
if 'province' in df.columns:
    # Top 10 provinces by average total enrollment
    province_enroll = df.groupby('province')['enrollment_total'].mean().sort_values(ascending=False).head(10)
    print("\nTop 10 provinces by average total enrollment:")
    print(province_enroll)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=province_enroll.values, y=province_enroll.index, palette='viridis')
    plt.title('Average Total Enrollment by Province (Top 10)')
    plt.xlabel('Average Enrollment')
    plt.show()

# ------------------------------
# 10. Correlation analysis
# ------------------------------
# Select numeric columns for correlation (excluding identifiers and high-cardinality)
corr_cols = ['enrollment_total', 'enrollment_girl', 'teachstaff_total', 'teachstaff_female',
             'totalstaff_total', 'totalstaff_female', 'grade_1_enrollment_total', 'grade_12_enrollment_total',
             't_primary', 't_lsec', 't_usec', 't_graduate']
corr_cols = [col for col in corr_cols if col in df.columns]
if len(corr_cols) > 1:
    corr_matrix = df[corr_cols].corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix of Key Variables')
    plt.show()

# ------------------------------
# 11. Pairplot for a subset of variables
# ------------------------------
pairplot_cols = ['enrollment_total', 'teachstaff_total', 'totalstaff_total', 'grade_1_enrollment_total']
pairplot_cols = [col for col in pairplot_cols if col in df.columns]
if len(pairplot_cols) >= 2:
    sns.pairplot(df)
    plt.suptitle('Pairplot of Selected Variables', y=1.02)
    plt.show()

# ------------------------------
# 12. Grade-level analysis: average repeaters by grade
# ------------------------------
grade_cols = [col for col in df.columns if col.startswith('grade_') and 'repeaters' in col]
if grade_cols:
    grade_repeaters = df[grade_cols].mean().sort_index()
    plt.figure(figsize=(12, 6))
    grade_repeaters.plot(kind='bar')
    plt.title('Average Number of Repeaters by Grade')
    plt.xlabel('Grade')
    plt.ylabel('Average Repeaters')
    plt.xticks(rotation=45)
    plt.show()

# ------------------------------
# 13. Promotion, repetition, dropout rates (if available)
# ------------------------------
rate_cols = [col for col in df.columns if '_promotion' in col or '_repetition' in col or '_dropout' in col]
if rate_cols:
    # Melt the rates for visualization
    rates_melted = df[rate_cols + ['type', 'year']].melt(id_vars=['type', 'year'], var_name='rate_type', value_name='rate')
    # Extract grade and rate category
    rates_melted['grade'] = rates_melted['rate_type'].str.extract(r'grade_(\d+)_').fillna(0).astype(int)
    rates_melted['rate_category'] = rates_melted['rate_type'].str.extract(r'(promotion|repetition|dropout)').fillna('')
    # Plot trends for a specific rate, e.g., promotion
    if 'promotion' in rates_melted['rate_category'].unique():
        promotion_data = rates_melted[rates_melted['rate_category'] == 'promotion']
        plt.figure(figsize=(14, 6))
        sns.lineplot(data=promotion_data, x='grade', y='rate', hue='type', style='year', markers=True, dashes=False)
        plt.title('Promotion Rates by Grade and School Type')
        plt.xlabel('Grade')
        plt.ylabel('Promotion Rate (%)')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()

# ------------------------------
# 14. Additional: Gender distribution
# ------------------------------
if 'enrollment_girl' in df.columns and 'enrollment_total' in df.columns:
    df['girl_ratio'] = df['enrollment_girl'] / df['enrollment_total']
    plt.figure(figsize=(10, 6))
    sns.histplot(df['girl_ratio'], bins=50, kde=True)
    plt.title('Distribution of Female Enrollment Ratio')
    plt.xlabel('Proportion of Girls')
    plt.show()

    # By province
    if 'province' in df.columns:
        province_girl_ratio = df.groupby('province')['girl_ratio'].mean().sort_values(ascending=False)
        plt.figure(figsize=(12, 8))
        province_girl_ratio.plot(kind='bar')
        plt.title('Average Female Enrollment Ratio by Province')
        plt.ylabel('Girls / Total Enrollment')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()

print("EDA completed.")
```
