```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
public_school_2018_2020 = pd.read_csv('public_cleaned.csv')
private_school_2018_2020 = pd.read_csv('private_cleaned.csv')
```

```python
public_school_2018_2020.insert(0,'type','public')
private_school_2018_2020.insert(0,'type','private')
```

```python
cols = public_school_2018_2020.columns.tolist()

# Print 5 columns per line
for i in range(0, len(cols), 5):
    print(f"{i:3}: {', '.join(cols[i:i+5])}")
```

```python
cols = private_school_2018_2020.columns.tolist()

# Print 5 columns per line
for i in range(0, len(cols), 5):
    print(f"{i:3}: {', '.join(cols[i:i+5])}")
```

```python
private_school_2018_2020=private_school_2018_2020.rename(columns ={"Unnamed: 1":"Year"})
```

```python
df = pd.concat([public_school_2018_2020,private_school_2018_2020],axis=0)
```

```python
df =df.drop(columns=['index'])
```

```python
df.isna().sum()
```

```python
df['Grade_1_Dropout'] = df['Grade_1_Dropout'].fillna(df['Grade_1_Dropout'].mean())
df['Grade_2_Dropout'] = df['Grade_2_Dropout'].fillna(df['Grade_2_Dropout'].mean())
df['Grade_3_Dropout'] = df['Grade_3_Dropout'].fillna(df['Grade_3_Dropout'].mean())
df['Grade_4_Dropout'] = df['Grade_4_Dropout'].fillna(df['Grade_4_Dropout'].mean())
df['Grade_5_Dropout'] = df['Grade_5_Dropout'].fillna(df['Grade_5_Dropout'].mean())
df['Grade_10_Dropout'] = df['Grade_10_Dropout'].fillna(df['Grade_10_Dropout'].mean())
df['Grade_11_Dropout'] = df['Grade_11_Dropout'].fillna(df['Grade_11_Dropout'].mean())
```

```python
df.set_index(['Province'], inplace=True)
```

```python
df.to_csv("final_clean.csv")
```
