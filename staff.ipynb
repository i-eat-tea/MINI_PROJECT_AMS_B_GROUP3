```python
def clean_sheet_v2(file_data, sheet_name):
    df = pd.read_excel(file_data, sheet_name=sheet_name, header=2)
    df = df.loc[:, ~df.columns.astype(str).str.contains('^Unnamed')]
    df = df.dropna(how='all')
    df.reset_index(drop=True, inplace=True)
    return df
```
