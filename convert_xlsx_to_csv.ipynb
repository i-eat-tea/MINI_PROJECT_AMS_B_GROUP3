# Convert Excel Files to CSV

Converts private and public school Excel files to CSV format.

```python
import pandas as pd
import os

OUTPUT_DIR = 'organized_education_data'
```

## Private Schools (specific sheet)

```python
PRIVATE_SHEET = 'Schools, Classes & Staff by Pro'

private_files = [
    ('2018-2019', f'{OUTPUT_DIR}/private_schools_2018-2019.xlsx'),
    ('2019-2020', f'{OUTPUT_DIR}/private_schools_2019-2020.xlsx'),
    ('2020-2021', f'{OUTPUT_DIR}/private_schools_2020-2021.xlsx'),
]

summary = []

for year, path in private_files:
    df = pd.read_excel(path, sheet_name=PRIVATE_SHEET)
    out_name = f'private_schools_{year}.csv'
    out_path = os.path.join(OUTPUT_DIR, out_name)
    df.to_csv(out_path, index=False)
    summary.append({'file': out_name, 'rows': len(df), 'cols': len(df.columns)})
    print(f'✅ {out_name} ({len(df)} rows × {len(df.columns)} cols)')
    print(f'   Columns: {list(df.columns)}\n')
```

## Public Schools (all sheets combined)

```python
public_files = [
    ('2018-2019', f'{OUTPUT_DIR}/public_schools_2018-2019.xlsx'),
    ('2019-2020', f'{OUTPUT_DIR}/public_schools_2019-2020.xlsx'),
    ('2020-2021', f'{OUTPUT_DIR}/public_schools_2020-2021.xlsx'),
]

for year, path in public_files:
    sheets = pd.read_excel(path, sheet_name=None)
    combined = pd.concat(sheets.values(), ignore_index=True)
    out_name = f'public_schools_{year}.csv'
    out_path = os.path.join(OUTPUT_DIR, out_name)
    combined.to_csv(out_path, index=False)
    summary.append({'file': out_name, 'rows': len(combined), 'cols': len(combined.columns)})
    print(f'✅ {out_name} ({len(combined)} rows × {len(combined.columns)} cols)')
    print(f'   Sheets merged: {list(sheets.keys())}')
    print(f'   Columns: {list(combined.columns)}\n')
```

## Conversion Summary

```python
print('=== Conversion Summary ===')
for s in summary:
    print(f"{s['file']:45s}  {s['rows']:5d} rows  {s['cols']:3d} cols")
```
