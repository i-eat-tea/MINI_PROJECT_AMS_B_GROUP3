```python
def drop_sparse(df, threshold=0.5):
    """Drop cols/rows where null ratio exceeds threshold, and duplicate columns."""
    df = df.loc[:, ~df.columns.duplicated()]           # duplicate cols
    df = df.loc[:, df.isnull().mean() < threshold]     # sparse cols
    df = df.loc[df.isnull().mean(axis=1) < threshold]  # sparse rows
    return df
```

```python
def diff_columns(df1, df2):
    cols1 = set(df1.columns)
    cols2 = set(df2.columns)
    only_in_df1 = cols1 - cols2
    only_in_df2 = cols2 - cols1
    in_both = cols1 & cols2
    
    print(f"only in df1:  {len(only_in_df1)}")
    print(f"only in df2:  {len(only_in_df2)}")
    print(f"in both:      {len(in_both)}")
    
    return {"only_in_df1": only_in_df1, "only_in_df2": only_in_df2, "in_both": in_both}
```

```python
from itertools import combinations

dfs = {
    "public16": public16,
    "public17": public17,
    "public18": public18,
    "public19": public19,
    "public20": public20,
    "public21": public21,
    "public22": public22,
    "public23": public23,
}

for (name1, df1), (name2, df2) in combinations(dfs.items(), 2):
    print(f"\n=== {name1} vs {name2} ===")
    
    diff_columns(df1, df2)
```
